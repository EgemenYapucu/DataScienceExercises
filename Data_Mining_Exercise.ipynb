```python
#Soru 9: Bir fonksiyon tanımlayın. Buradaki amaç location y, ve location x verilerini
#fonksiyona verdiğinizde, şutun atıldığı noktanın potaya kuşbakışı uzaklığını(feet)
#buldurmaktır. Bu uzaklıkları 0-10,10-20,20-30, 30+ olacak şekilde,
#yakın, orta, uzak ve orta sahadan gibi isimlerle sınıflandırın.
#Bu sınıflandırmayı verinize yeni bir sütun(feature) olarak ekleyin.

def distance(x, y):
    dist = ((np.sqrt(np.square(x)+np.square(y)))/100)/0.3048#x ile y verisinin karesi alınıp toplandılar ardından karekök alınıp gerekli sabit ile çarpıldı.
    return dist

locx = df.iloc[:,4]
locy = df.iloc[:,10]

feature = distance(locx,locy)

df['Distance'] = feature#Elde edilen sayılar için Distance sütunu oluşturuldu

for i in range(len(feature)):
    if feature[i] <= 10:
        feature[i] = 'yakın'
    elif feature[i] <= 20:
        feature[i] = 'orta'
    elif feature[i] <= 30:
        feature[i] = 'uzak'
    else:
        feature[i] = 'orta sahadan'

df['Feature'] = feature#Elde edilen elemanlar için Feature sütunu oluşturuldu
df['Distance']
```
